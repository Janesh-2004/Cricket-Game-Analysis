```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data =pd.read_csv('E:\\ODI_Match_info.csv')
df = pd.DataFrame(data)
cross_tab = pd.crosstab(df['team1'], df['winner'])
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Match Outcomes by Teams')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('E:\\ODI_Match_info.csv')
df = pd.DataFrame(data)
decision_counts = df.groupby(['team1', 'toss_decision']).size().unstack(fill_value=0)
decision_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Toss Decisions by Teams')
plt.xlabel('Team')
plt.ylabel('Number of Toss Decisions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('E:\\ODI_Match_info.csv')
df = pd.DataFrame(data)
df['win'] = df['team1'] == df['winner']
win_percentage = df.groupby('season')['win'].mean() * 100
win_percentage.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Team Win Percentage Over Seasons')
plt.xlabel('Season')
plt.ylabel('Win Percentage (%)')
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv('E:\\ODI_Match_info.csv')
df = pd.DataFrame(data)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['win_by_runs'], df['win_by_wickets'], df['id'], c='blue', marker='o')
ax.set_xlabel('Win by Runs')
ax.set_ylabel('Win by Wickets')
ax.set_zlabel('id')
plt.title('3D Scatter Plot for Multivariate Analysis')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
data =  pd.read_csv('E:\\ODI_Match_info.csv')
df = pd.DataFrame(data)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x = df['win_by_runs']
y = df['win_by_wickets']
z = df['id']
X, Y = np.meshgrid(x, y)
Z = X + Y
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('Win by Runs')
ax.set_ylabel('Win by Wickets')
ax.set_zlabel('id')
plt.title('3D Surface Plot of Cricket Match Data')
plt.show()
```
